```python
# Using the csv library to read csv files
import json
import csv
import pandas as pd
import numpy as np

rows = []

with open('all_places_of_worship.csv', 'r+') as f:
    
    csvreader = csv.reader(f)
    header = next(csvreader)
    
    for row in csvreader:
        rows.append(row)

print("Header:" + str(header))
print("First Row: " + str(rows[0]))
print("Number of rows: " + str(len(rows)))
```

```python
# Creating a DataFrame object to clean duplicates, Strategy 1
df_duplicate = df
  
# Selecting duplicate rows except last occurrence based on all columns.
duplicate = df_duplicate[df_duplicate.duplicated(keep = 'last')]
  
print("Duplicate Rows :")
  
# Print the resultant Dataframe
duplicate
```

```python
# Creating a DataFrame object to clean duplicates, Strategy 2
# To clean basis name, address, city & state

df_selective = df_duplicate[['name','address','city','state']]

duplicate = df_selective[df_selective.duplicated(keep = 'last')]
  
print("Duplicate Rows :")
  
# Print the resultant Dataframe
duplicate
```
